```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
# import csv file

df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
df.head(6)
```

```python
df.shape
```

```python
df.info()
```

```python
#drop unrelated/blank columns

df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
df.head(6)
```

```python
#check for null values

pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape  #check
```

```python
# change data type

df['Amount'] = df['Amount'].astype('int')
```

<h1 style="color:red;">Exploratory Data Analysis</h1>

<h4 style="color:red">Gender</h4>

```python
plt.figure(figsize=(6, 3.5))  # Set width and height in inches

ax = sns.countplot(x = 'Gender',data = df,width=0.5 ,hue='Gender', palette='viridis',legend=False)

for bars in ax.containers:
    ax.bar_label(bars)
# Adding labels and title
plt.title('Count of Order by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.tight_layout()
sns.despine()
plt.show()
```

<h4 style="color:blue">This shows that a significant portion of our customer base is female.</h4>

```python
# Grouping and sorting
sales_gen = df.groupby('Gender', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Plotting
plt.figure(figsize=(7, 3.5))
sns.barplot(x='Gender', y='Amount', data=sales_gen,hue='Gender',width=0.5, palette='viridis',legend=False)

# Adding labels and title
plt.title('Total Sales Amount by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Amount')
plt.tight_layout()
sns.despine()
plt.show()
```

<h4 style="color:blue">This indicates that the total amount from female customers is higher.</h4>

<h4 style="color:red">AGE</h4>

```python
# Plotting
plt.figure(figsize=(7, 3.5))
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender',width=0.5, palette='viridis',legend=True)

for bars in ax.containers:
    ax.bar_label(bars)
# Adding labels and title
plt.title('Total Sales Amount by Gender')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.tight_layout()
sns.despine()
plt.show()
```

<h4 style="color:blue">This shows that a significant portion of our customer base from 26-35 age gr.</h4>

```python
sales_gen = df.groupby('Age Group', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Plotting
plt.figure(figsize=(7, 3.5))
sns.barplot(x='Age Group', y='Amount', data=sales_gen,hue='Age Group',width=0.5, palette='viridis',legend=False)

# Adding labels and title
plt.title('Total Sales Amount by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Amount')
plt.tight_layout()
sns.despine()
plt.show()
```

<h4 style="color:blue">This show that highest amount of shopping done by 26-35 age group.</h4>

<h1 style="color:red">State wise analysis</h1>

```python
sales_gen = df.groupby('State', as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(12)

# Plotting
plt.figure(figsize=(13,3.5))
sns.barplot(x='State', y='Orders', data=sales_gen,hue='State',width=0.5, palette='viridis',legend=False)

# Adding labels and title
plt.title('Number of orders by State')
plt.xticks(rotation=45)
plt.xlabel('State')
plt.ylabel('Order')
plt.tight_layout()
sns.despine()
plt.show()
```

<h4 style="color:blue">This graph show number of orders state-wise. In which, we found out we are getting most of orders from Uttar Pradesh.</h4>

```python
sales_gen = df.groupby('State', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(12)

# Plotting
plt.figure(figsize=(13,3.5))
sns.barplot(x='State', y='Amount', data=sales_gen,hue='State',width=0.5, palette='viridis',legend=False)

# Adding labels and title
plt.title('Total Amount of sales by State')
plt.xlabel('State')
plt.ylabel('Total Amount')
plt.xticks(rotation=45)
plt.tight_layout()
sns.despine()
plt.show()
```

<h4 style="color:blue">This graph shows total revenue generated state-wise. In which, we found out we are getting highest revenue from Uttar Pradesh followeed by maharashtra, karnataka, and Delhi.</h4>

<h1 style="color:red">Occupation wise analysis-</h1>

```python
# Calculate order of occupations by count
order = df['Occupation'].value_counts().index

plt.figure(figsize=(12, 3.5))
ax = sns.countplot(data=df, x='Occupation', hue='Gender', width=0.4,
                   palette='viridis', legend=False, order=order)

# Add value labels
for bars in ax.containers:
    ax.bar_label(bars)

# Labels and title
plt.title('Number of Orders by Occupation')
plt.xticks(rotation=45)
plt.tight_layout()
sns.despine()
plt.show()
```

<h4 style="color:blue">This graph shows that most of the orders are coming from the IT sector, followed by healthcare, aviation, banking, and government sectors, with a higher prevalence of female customers across these industries.</h4>

```python
sales_gen = df.groupby('Occupation', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Plotting
plt.figure(figsize=(13, 4))
sns.set_style("whitegrid")
sns.barplot(x='Occupation', y='Amount', data=sales_gen ,hue='Occupation',width=0.4, palette='viridis',legend=False)

# Adding labels and title
plt.title('Total Amount of sales by Occupation')
plt.xticks(rotation=45)
plt.tight_layout()
sns.despine()
plt.show()
```

<h4 style="color:blue">This graph shows that most of the revenue are coming from the IT sector, followed by healthcare, aviation, banking, and government sectors.</h4>

<h1 style="color:red">Product Category wise analysis-</h1>

```python
# Calculate order of occupations by count
order = df['Product_Category'].value_counts().index

plt.figure(figsize=(12, 3.5))
ax = sns.countplot(data=df, x='Product_Category', hue='Product_Category', width=0.4,
                   palette='viridis', legend=False, order=order)

# Add value labels
for bars in ax.containers:
    ax.bar_label(bars)

# Labels and title
plt.title('Number of Orders by Occupation')
plt.xticks(rotation=45)
plt.tight_layout()
sns.despine()
plt.show()
```

<h4 style="color:blue">This graph shows that most shopped category is Clothing & Apparel followed by Food and Electronics.</h4>

```python
sales_gen = df.groupby('Product_Category', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Plotting
plt.figure(figsize=(13, 4))
sns.barplot(x='Product_Category', y='Amount', data=sales_gen ,hue='Product_Category',width=0.4, palette='viridis',legend=False)

# Adding labels and title
plt.title('Total Amount of sales by Product_Category')
plt.xticks(rotation=45)
plt.tight_layout()
sns.despine()
plt.show()
```

<h4 style="color:blue">From above graphs we can see that most revenue generated from Food, Clothing and Electronics category</h4>

```python
sales_gen = df.groupby('Product_ID', as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

# Plotting
plt.figure(figsize=(13, 4))
sns.set_style("whitegrid")
sns.barplot(x='Product_ID', y='Orders', data=sales_gen ,hue='Product_ID',width=0.4, palette='viridis',legend=False)

# Adding labels and title
plt.title('Total Orders by Product_ID')
plt.xticks(rotation=45)
plt.tight_layout()
sns.despine()
plt.show()
```
